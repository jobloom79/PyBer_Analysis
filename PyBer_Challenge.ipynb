### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# Read the City and Ride Data
city_data_df = pd.read_csv("Resources/city_data.csv")
ride_data_df = pd.read_csv("Resources/ride_data.csv")
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides=pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# city_data_df.head()
# ride_data_df.head()
```

```python
# 2. Get the total drivers for each total type
total_drivers=city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each total type
total_fares=pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
avg_ride_fare=pyber_data_df.groupby(["type"]).sum()["fare"]/pyber_data_df.groupby(["type"]).count()["ride_id"]
avg_ride_fare
```

```python
# 5. Get the average fare per driver for each driver type. 
avg_driver_fare=pyber_data_df.groupby(["type"]).sum()["fare"]/city_data_df.groupby(["type"]).sum()["driver_count"]
avg_driver_fare
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df=pd.DataFrame(
    {"Total Rides":total_rides,
    "Total Drivers":total_drivers,
     "Total Fares":total_fares,
    "Average Fare per Ride":avg_ride_fare,
    "Average Fare per Driver":avg_driver_fare
    })
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df=pd.DataFrame(
    {"Total Rides":total_rides,
    "Total Drivers":total_drivers,
     "Total Fares":total_fares,
    "Average Fare per Ride":avg_ride_fare,
    "Average Fare per Driver":avg_driver_fare
    })
pyber_summary_df["Total Rides"]=pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"]=pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"]=pyber_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pyber_summary_df["Average Fare per Driver"]=pyber_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
    #  Using groupby() to create a new DataFrame showing the sum of the fares 
    #  for each date where the indices are the city type and date.
pyber_total_df = pyber_data_df.groupby(["date","type"]).sum()[["fare"]]
pyber_total_df
```

```python
# 2. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
pyber_total_df=pyber_total_df.reset_index()
pyber_total_df.head(10)
```

```python
# 3. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_total_pivot = pyber_total_df.pivot(index="date", columns="type", values="fare")
pyber_total_pivot
```

```python
# 4. Create new dataframe using loc
pyber_daterange_df=pyber_total_df.loc[(pyber_total_df['date']>='2019-01-01')&(pyber_total_df['date']<='2019-04-28')]
pyber_daterange_df
```

```python
# 4a. recreate pivot with new date range
pyber_daterange_pivot = pyber_daterange_df.pivot(index="date", columns="type", values="fare")
pyber_daterange_pivot
```

```python
# 5. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
pyber_daterange_pivot.index = pd.to_datetime(pyber_daterange_pivot.index)
```

```python
# 6. Check that the datatype for the index is datetime using df.info()
pyber_daterange_pivot.info()
```

```python
# 7. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_pivot_df_week = pyber_daterange_pivot.resample('W').sum()
pyber_pivot_df_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

pyber_pivot_df_week.plot()
plt.xlabel("")
plt.ylabel("Fare($USD)")
plt.show()
```
